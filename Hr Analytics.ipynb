```python
sb.countplot(x='Attrition', data=df)
plt.show()
```

##### The bar chart compares employees who have left (Attrition: Yes) with those who have stayed (Attrition: No).

##### Most employees stay with the company.

##### Fewer employees left the company.

```python
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
sb.countplot(x='Gender', hue='Attrition', data=df)


plt.subplot(1,2,2)
sb.countplot(x='MaritalStatus', hue='Attrition', data=df, )

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
sb.countplot(x='EducationField', hue='Attrition', data=df)
plt.xticks(rotation=90)


plt.subplot(2,2,2)
sb.barplot(x='Attrition', y='Age', data=df)
plt.show()
```

##### Gender & Attrition: More males are leaving the company compared to females.

##### Marital Status & Attrition: Single employees tend to leave more compared to married and divorced employees.Married employees are leaving the company compared to single and divorced employees.

##### Education Field & Attrition: Employees in Life Sciences and Medical fields have the highest attrition, while fewer employees are leaving in fields like Human Resources and Technical Degree.

##### Age & Attrition: Employees who leave tend to be younger on average compared to those who stay.

```python
plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
sb.boxplot(x='Attrition', y='JobSatisfaction', data=df)

plt.subplot(1,3,2)
sb.boxplot(x='Attrition', y='MonthlyIncome', data=df)

plt.subplot(1,3,3)
sb.boxplot(x='Attrition', y='PerformanceRating', data=df)

plt.show()
```

##### job Satisfaction & Attrition: Employees who left (Attrition = Yes) generally had lower job satisfaction compared to those who stayed. 

##### Monthly Income & Attrition: Employees who stayed (Attrition = No) had a higher median income, while those who left had lower salaries. 
    
##### Performance Rating & Attrition: There is no difference in performance ratings between those who left and those who stayed.
